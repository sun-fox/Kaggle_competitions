```python
import pandas as pd
```

```python
train = pd.read_table("train.csv",sep=",")
train
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
train.describe()
```

```python
Y = train.RMSD
X = train.drop(['RMSD'],axis=1)

import matplotlib.pyplot as plt
plt.plot(X,train.RMSD)
plt.xlabel('X-axis')
plt.ylabel('RMSD')
```

```python
lr = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
lr.fit(X,Y)
tst = pd.read_csv("test.csv")
```

```python
print(lr)
tst
```

```python
X
```

```python
ans = lr.predict(tst)
```

```python
from sklearn.metrics import r2_score
score=r2_score(train.RMSD,ans)
```
